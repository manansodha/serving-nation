## Implementation Details

### Trust Scoring Algorithm

```python
# Rule-Based Penalties
if claims_specialties AND equipment == '[]':
    penalty += 30  # Claims without equipment

if facilityType == 'medical' AND doctor_count == 0:
    penalty += 25  # No doctors

if officialPhone == null AND officialEmail == null:
    penalty += 20  # No contact

if website == null:
    penalty += 15  # No online presence

Trust_Score = 100 - penalty
```

### Validation Penalties (Agent-Based)

```python
# Medical Standards Cross-Reference
if claims('surgery') AND missing('operating room'):
    validation_penalty += 30

if claims('oncology') AND missing('chemotherapy equipment'):
    validation_penalty += 30

if claims('emergency') AND missing('defibrillator'):
    validation_penalty += 35

Final_Trust_Score = Trust_Score - Validation_Penalty
```

### H3 Geospatial Queries

```sql
-- Regional capacity aggregation
SELECT 
    h3_index_res7,
    COUNT(*) as facility_count,
    AVG(Final_Trust_Score) as avg_trust,
    COUNT(CASE WHEN 'emergency' IN specialties THEN 1 END) as emergency_centers
FROM gold_healthcare_facilities_geo
GROUP BY h3_index_res7
HAVING emergency_centers > 0
```

````python
from pyspark.sql.functions import col, when, lit, coalesce, size, from_json, array_contains
from pyspark.sql.types import ArrayType, StringType

# Rule-Based Validation: Catch obvious lies and inconsistencies
df_evaluated = df.withColumn(
    "Trust_Penalty",
    # Initialize to 0
    lit(0)
).withColumn(
    "Trust_Penalty",
    # Rule 1: Claims specialties but has no equipment listed
    when(
        (col("specialties").isNotNull()) & (col("specialties") != "null") & 
        (col("specialties") != "[]") &
        ((col("equipment").isNull()) | (col("equipment") == "null") | (col("equipment") == "[]")),
        col("Trust_Penalty") + 30  # High risk: claims specialties without equipment
    ).otherwise(col("Trust_Penalty"))
).withColumn(
    "Trust_Penalty",
    # Rule 2: No doctors listed but claims to be a clinic/hospital
    when(
        ((col("numberDoctors").isNull()) | (col("numberDoctors") == "null") | (col("numberDoctors") == "0")) &
        (col("facilityTypeId").isin("clinic", "hospital")),
        col("Trust_Penalty") + 25  # Red flag: medical facility with no doctors
    ).otherwise(col("Trust_Penalty"))
).withColumn(
    "Trust_Penalty",
    # Rule 3: No contact information (phone, email, website)
    when(
        ((col("officialPhone").isNull()) | (col("officialPhone") == "null")) &
        ((col("email").isNull()) | (col("email") == "null")) &
        ((col("officialWebsite").isNull()) | (col("officialWebsite") == "null")),
        col("Trust_Penalty") + 20  # Suspicious: no way to contact facility
    ).otherwise(col("Trust_Penalty"))
).withColumn(
    "Trust_Penalty",
    # Rule 4: No social media presence and no website
    when(
        ((col("officialWebsite").isNull()) | (col("officialWebsite") == "null")) &
        ((col("facebookLink").isNull()) | (col("facebookLink") == "null")) &
        ((col("twitterLink").isNull()) | (col("twitterLink") == "null")) &
        ((col("linkedinLink").isNull()) | (col("linkedinLink") == "null")) &
        ((col("instagramLink").isNull()) | (col("instagramLink") == "null")),
        col("Trust_Penalty") + 15  # Low online presence
    ).otherwise(col("Trust_Penalty"))
).withColumn(
    "Trust_Penalty",
    # Cap penalty at 100
    when(col("Trust_Penalty") > 100, lit(100)).otherwise(col("Trust_Penalty"))
)

# Register as temp view for SQL access
df_evaluated.createOrReplaceTempView("temp_evaluated")

print(f"Rule-based validation complete. Penalty scores range from 0-100.")
display(df_evaluated.select("name", "Trust_Penalty", "specialties", "equipment").limit(10))
````

````python
%sql
-- Create Gold Delta Table with Trust Scoring
CREATE OR REPLACE TABLE gold_healthcare_facilities AS
SELECT 
    *,
    -- Trust Score: 100 (perfect) minus penalties
    (100 - Trust_Penalty) AS Trust_Score,
    
    -- Trust Reason: Explain why penalties were applied
    CONCAT_WS('; ',
        CASE 
            WHEN specialties IS NOT NULL AND specialties != 'null' AND specialties != '[]'
                 AND (equipment IS NULL OR equipment = 'null' OR equipment = '[]')
            THEN 'Claims specialties without listing equipment'
            ELSE NULL
        END,
        CASE
            WHEN (numberDoctors IS NULL OR numberDoctors = 'null' OR numberDoctors = '0')
                 AND facilityTypeId IN ('clinic', 'hospital')
            THEN 'Medical facility with no doctors listed'
            ELSE NULL
        END,
        CASE
            WHEN (officialPhone IS NULL OR officialPhone = 'null')
                 AND (email IS NULL OR email = 'null')
                 AND (officialWebsite IS NULL OR officialWebsite = 'null')
            THEN 'No contact information available'
            ELSE NULL
        END,
        CASE
            WHEN (officialWebsite IS NULL OR officialWebsite = 'null')
                 AND (facebookLink IS NULL OR facebookLink = 'null')
                 AND (twitterLink IS NULL OR twitterLink = 'null')
                 AND (linkedinLink IS NULL OR linkedinLink = 'null')
                 AND (instagramLink IS NULL OR instagramLink = 'null')
            THEN 'No online presence'
            ELSE NULL
        END
    ) AS Trust_Reason
    
    -- LLM-as-a-Judge (Cross-referencing) - COMMENTED OUT until model endpoint is deployed
    -- Uncomment and replace 'YOUR_MODEL_ENDPOINT' with actual endpoint name:
    /*
    ,ai_query(
        'YOUR_MODEL_ENDPOINT',  -- e.g., 'databricks-meta-llama-3-3-70b-instruct' or your Foundation Model API endpoint
        CONCAT(
            'Audit this healthcare facility for contradictions. ',
            'Specialties claimed: ', COALESCE(specialties, 'none'), '. ',
            'Equipment listed: ', COALESCE(equipment, 'none'), '. ',
            'Capabilities: ', COALESCE(capability, 'none'), '. ',
            'Does the equipment physically support the claimed specialties and capabilities? ',
            'Answer ONLY with: "Risk_Score|Reason" where Risk_Score is 0-50 and Reason is one sentence.'
        )
    ) AS LLM_Audit_Result
    */
FROM temp_evaluated;

-- Show summary of trust scores
SELECT 
    CASE 
        WHEN Trust_Score >= 90 THEN 'Highly Trusted (90-100)'
        WHEN Trust_Score >= 70 THEN 'Trusted (70-89)'
        WHEN Trust_Score >= 50 THEN 'Moderate Risk (50-69)'
        ELSE 'High Risk (0-49)'
    END AS Trust_Category,
    COUNT(*) AS Facility_Count
FROM gold_healthcare_facilities
GROUP BY Trust_Category
ORDER BY MIN(Trust_Score) DESC;
````

````python
%sql
-- View examples of facilities with their Trust Scores and Reasons
SELECT 
    name,
    facilityTypeId,
    Trust_Score,
    Trust_Penalty,
    Trust_Reason,
    CASE 
        WHEN Trust_Score >= 90 THEN 'Highly Trusted'
        WHEN Trust_Score >= 70 THEN 'Trusted'
        WHEN Trust_Score >= 50 THEN 'Moderate Risk'
        ELSE 'High Risk'
    END AS Risk_Level
FROM gold_healthcare_facilities
WHERE Trust_Reason IS NOT NULL AND Trust_Reason != ''
ORDER BY Trust_Penalty DESC
LIMIT 20;
````

##  Vector Search +  Geospatial Enhancement

### Vector Search Benefits
- **Semantic Matching**: Find facilities by natural language descriptions ("emergency trauma care", "pediatric specialists")
- **Auto-sync**: Vector index automatically updates as the Gold table changes
- **Complex Queries**: Match based on meaning, not just keywords

### H3 Geospatial Benefits
- **Ultra-Fast Distance Calculations**: H3 indices enable spatial queries in milliseconds
- **Specialized Deserts**: Identify regions lacking high-acuity or specialty care
- **Coverage Analysis**: Calculate facility density and accessibility metrics
- **Hexagonal Grid**: H3 provides uniform spatial distribution analysis

````python
%sql
-- Add H3 Geospatial Indices to Gold Table
-- H3 provides hexagonal spatial indexing for fast distance calculations

CREATE OR REPLACE TABLE gold_healthcare_facilities_geo AS
SELECT 
    *,
    -- H3 indices at different resolutions (8 = ~0.5km, 7 = ~5km, 6 = ~36km)
    h3_longlatash3(CAST(longitude AS DOUBLE), CAST(latitude AS DOUBLE), 8) AS h3_index_res8,
    h3_longlatash3(CAST(longitude AS DOUBLE), CAST(latitude AS DOUBLE), 7) AS h3_index_res7,
    h3_longlatash3(CAST(longitude AS DOUBLE), CAST(latitude AS DOUBLE), 6) AS h3_index_res6
FROM gold_healthcare_facilities
WHERE latitude IS NOT NULL 
  AND longitude IS NOT NULL 
  AND latitude != 'null' 
  AND longitude != 'null'
  AND TRY_CAST(latitude AS DOUBLE) IS NOT NULL
  AND TRY_CAST(longitude AS DOUBLE) IS NOT NULL;

-- Show distribution
SELECT 
    COUNT(*) as facilities_with_location,
    COUNT(DISTINCT h3_index_res8) as unique_h3_cells_fine,
    COUNT(DISTINCT h3_index_res7) as unique_h3_cells_medium,
    COUNT(DISTINCT h3_index_res6) as unique_h3_cells_coarse
FROM gold_healthcare_facilities_geo;
````

````python
%sql
-- Prepare data for Vector Search embeddings
-- Create a rich text representation of each facility for semantic search

CREATE OR REPLACE TABLE gold_healthcare_facilities_vector_ready AS
SELECT 
    *,
    -- Combine key fields into searchable text profile
    CONCAT_WS(' | ',
        CONCAT('Facility: ', name),
        CONCAT('Type: ', COALESCE(facilityTypeId, 'unknown')),
        CONCAT('Location: ', COALESCE(address_city, ''), ', ', COALESCE(address_stateOrRegion, ''), ', ', COALESCE(address_country, '')),
        CONCAT('Specialties: ', COALESCE(specialties, 'none')),
        CONCAT('Equipment: ', COALESCE(equipment, 'none')),
        CONCAT('Capabilities: ', COALESCE(capability, 'none')),
        CONCAT('Description: ', COALESCE(description, 'none')),
        CONCAT('Trust Score: ', CAST(Trust_Score AS STRING), '/100'),
        CASE WHEN Trust_Reason IS NOT NULL AND Trust_Reason != '' 
             THEN CONCAT('Trust Issues: ', Trust_Reason)
             ELSE 'No trust issues'
        END
    ) AS facility_profile_text
FROM gold_healthcare_facilities_geo;

-- Preview the searchable profiles
SELECT 
    name,
    Trust_Score,
    LEFT(facility_profile_text, 200) AS profile_preview
FROM gold_healthcare_facilities_vector_ready
LIMIT 5;
````

````python
%sql
-- Identify "Specialized Deserts" - regions lacking high-quality specialty care
-- Using H3 spatial indices for fast geospatial analysis

WITH facility_coverage AS (
    SELECT 
        h3_index_res7 AS region,
        COUNT(*) AS total_facilities,
        COUNT(CASE WHEN Trust_Score >= 70 THEN 1 END) AS trusted_facilities,
        COUNT(CASE WHEN specialties IS NOT NULL AND specialties != 'null' AND specialties != '[]' THEN 1 END) AS specialty_facilities,
        COUNT(CASE WHEN facilityTypeId = 'hospital' THEN 1 END) AS hospitals,
        -- Check for high-acuity capabilities
        COUNT(CASE WHEN specialties LIKE '%surgery%' OR specialties LIKE '%Surgery%' THEN 1 END) AS surgical_facilities,
        COUNT(CASE WHEN specialties LIKE '%emergency%' OR specialties LIKE '%Emergency%' THEN 1 END) AS emergency_facilities,
        COUNT(CASE WHEN specialties LIKE '%trauma%' OR specialties LIKE '%Trauma%' THEN 1 END) AS trauma_facilities,
        -- Geographic context
        FIRST(address_stateOrRegion) AS state,
        FIRST(address_country) AS country
    FROM gold_healthcare_facilities_geo
    GROUP BY h3_index_res7
),
desert_analysis AS (
    SELECT 
        *,
        -- Calculate desert risk score (higher = more concerning)
        CASE 
            WHEN total_facilities = 0 THEN 100
            WHEN hospitals = 0 AND emergency_facilities = 0 THEN 80
            WHEN surgical_facilities = 0 AND trauma_facilities = 0 THEN 60
            WHEN trusted_facilities = 0 THEN 50
            WHEN specialty_facilities < 2 THEN 30
            ELSE 0
        END AS desert_risk_score
    FROM facility_coverage
)
SELECT 
    region AS h3_cell,
    state,
    country,
    total_facilities,
    trusted_facilities,
    hospitals,
    surgical_facilities,
    emergency_facilities,
    trauma_facilities,
    desert_risk_score,
    CASE 
        WHEN desert_risk_score >= 80 THEN 'Critical Desert'
        WHEN desert_risk_score >= 50 THEN 'High Risk'
        WHEN desert_risk_score >= 30 THEN 'Moderate Gap'
        ELSE 'Adequate Coverage'
    END AS coverage_status
FROM desert_analysis
WHERE desert_risk_score > 0  -- Only show areas with gaps
ORDER BY desert_risk_score DESC, total_facilities ASC
LIMIT 30;
````

````python
%sql
-- Fast Distance Calculations: Find nearby facilities using H3
-- Example: Find all facilities within ~5km radius of New Delhi

-- Sample location: New Delhi area
WITH target_location AS (
    SELECT 
        28.6139 AS target_lat,
        77.2090 AS target_lon,
        h3_longlatash3(77.2090, 28.6139, 7) AS target_h3
),
nearby_facilities AS (
    SELECT 
        f.name,
        f.facilityTypeId,
        f.Trust_Score,
        f.address_line1,
        f.address_city,
        f.address_stateOrRegion,
        f.specialties,
        f.h3_index_res7,
        t.target_h3,
        t.target_lat,
        t.target_lon,
        CAST(f.latitude AS DOUBLE) AS facility_lat,
        CAST(f.longitude AS DOUBLE) AS facility_lon
    FROM gold_healthcare_facilities_geo f
    CROSS JOIN target_location t
    WHERE f.h3_index_res7 IN (
        -- Get target cell and its immediate neighbors
        SELECT EXPLODE(h3_kring(t.target_h3, 2)) FROM target_location t
    )
),
with_distances AS (
    SELECT 
        *,
        -- Calculate precise distance using Haversine formula
        ST_DISTANCE(
            ST_POINT(target_lon, target_lat),
            ST_POINT(facility_lon, facility_lat)
        ) / 1000 AS distance_km
    FROM nearby_facilities
)
SELECT 
    name,
    facilityTypeId,
    Trust_Score,
    ROUND(distance_km, 2) AS distance_km,
    CONCAT(address_line1, ', ', address_city, ', ', address_stateOrRegion) AS address,
    LEFT(specialties, 100) AS specialties_preview
FROM with_distances
WHERE distance_km <= 10  -- Refine to exact 10km radius
ORDER BY distance_km ASC
LIMIT 20;
````

````python
# Step 3: Multi-Agent Validation System with Self-Correction
# Uses LangGraph to orchestrate Extractor → Validator → Correction loop

import json
from typing import TypedDict, Annotated
from langgraph.graph import StateGraph, END
import mlflow

# Enable MLflow tracing for full observability
mlflow.langchain.autolog()

print("MLflow auto-logging enabled - traces will appear in Databricks UI")

# Define the state that flows between agents
class FacilityValidationState(TypedDict):
    facility_name: str
    source_document_id: str
    raw_data: dict
    extracted_profile: dict
    validation_results: dict
    citations: list
    trust_score: int
    trust_reason: str
    iteration_count: int
    max_iterations: int

# Agent 1: Extractor
def extractor_agent(state: FacilityValidationState) -> FacilityValidationState:
    """
    Parses raw facility data and outputs structured JSON profile with citations.
    In production, this would call an LLM with citation-enforcing prompts.
    """
    print(f"\n[EXTRACTOR] Processing: {state['facility_name']}")
    
    raw = state['raw_data']
    
    # Extract claimed specialties
    specialties = raw.get('specialties', '')
    if specialties and specialties not in ['null', '[]']:
        # Parse specialties (simplified - in production use LLM)
        specialty_list = []
        if 'surgery' in specialties.lower():
            specialty_list.append('surgery')
        if 'emergency' in specialties.lower():
            specialty_list.append('emergency')
        if 'trauma' in specialties.lower():
            specialty_list.append('trauma')
        if 'oncology' in specialties.lower():
            specialty_list.append('oncology')
        if 'cardiology' in specialties.lower():
            specialty_list.append('cardiology')
        if 'pediatrics' in specialties.lower() or 'pediatric' in specialties.lower():
            specialty_list.append('pediatrics')
        if 'dentistry' in specialties.lower() or 'dental' in specialties.lower():
            specialty_list.append('dentistry')
    else:
        specialty_list = []
    
    # Extract equipment
    equipment_str = raw.get('equipment', '')
    equipment_list = [] if equipment_str in ['null', '[]', None] else [equipment_str]
    
    # Create structured profile with citations
    profile = {
        "facility_name": state['facility_name'],
        "claimed_specialties": specialty_list,
        "listed_equipment": equipment_list,
        "facility_type": raw.get('facilityTypeId', 'unknown'),
        "has_contact_info": bool(raw.get('officialPhone') and raw.get('officialPhone') != 'null')
    }
    
    # Add citations
    citations = [
        {
            "field": "specialties",
            "value": specialties,
            "source_document_id": state['source_document_id'],
            "page_number": 1,
            "raw_text": f"Specialties: {specialties}"
        },
        {
            "field": "equipment",
            "value": equipment_str,
            "source_document_id": state['source_document_id'],
            "page_number": 1,
            "raw_text": f"Equipment: {equipment_str}"
        }
    ]
    
    state['extracted_profile'] = profile
    state['citations'] = citations
    
    print(f"   Extracted profile: {len(specialty_list)} specialties claimed")
    return state

# Agent 2: Validator
def validator_agent(state: FacilityValidationState) -> FacilityValidationState:
    """
    Cross-references extracted profile against medical ground truth standards.
    Identifies missing requirements and calculates validation penalties.
    """
    print(f"\n[🔍 VALIDATOR] Validating against medical standards...")
    
    profile = state['extracted_profile']
    violations = []
    validation_penalty = 0
    
    # Load ground truth standards (in production, query from Delta table)
    # For demo, using hardcoded critical requirements
    standards = {
        'surgery': ['anesthesiologist', 'sterilization unit', 'operating room'],
        'emergency': ['emergency physician', '24/7 operation'],
        'trauma': ['trauma surgeon', 'ICU'],
        'oncology': ['oncologist', 'biohazard disposal'],
        'cardiology': ['cardiologist', 'ECG machine'],
        'pediatrics': ['pediatrician'],
        'dentistry': ['sterilization equipment', 'licensed dentist']
    }
    
    # Validate each claimed specialty
    for specialty in profile['claimed_specialties']:
        if specialty in standards:
            required_items = standards[specialty]
            equipment_str = ' '.join(profile['listed_equipment']).lower()
            
            for requirement in required_items:
                # Check if requirement is mentioned in equipment
                if requirement.lower() not in equipment_str:
                    violations.append({
                        "specialty": specialty,
                        "missing_requirement": requirement,
                        "severity": "critical",
                        "penalty_points": 30
                    })
                    validation_penalty += 30
                    print(f"Missing {requirement} for {specialty}")
    
    # Check if facility type matches claimed specialties
    if profile['claimed_specialties'] and profile['facility_type'] not in ['hospital', 'clinic']:
        violations.append({
            "issue": "Facility type mismatch",
            "detail": f"Claims specialties but facility_type is '{profile['facility_type']}'",
            "severity": "high",
            "penalty_points": 20
        })
        validation_penalty += 20
    
    state['validation_results'] = {
        "violations": violations,
        "validation_penalty": validation_penalty,
        "total_violations": len(violations)
    }
    
    print(f"  Found {len(violations)} violations, penalty: {validation_penalty} points")
    return state

# Decision node: Should we correct?
def should_correct(state: FacilityValidationState) -> str:
    """
    Decides whether to loop back for correction or proceed to final scoring.
    """
    violations = state['validation_results']['violations']
    iteration = state['iteration_count']
    max_iter = state['max_iterations']
    
    # If critical violations found and haven't exceeded max iterations
    critical_violations = [v for v in violations if v.get('severity') == 'critical']
    
    if critical_violations and iteration < max_iter:
        print(f"\n[CORRECTION] {len(critical_violations)} critical violations - looping back")
        return "correct"
    else:
        print(f"\n[FINAL] Proceeding to final scoring")
        return "finalize"

# Correction node
def correction_agent(state: FacilityValidationState) -> FacilityValidationState:
    """
    Applies corrections based on validation failures.
    In production, this would re-query the LLM with validation feedback.
    """
    print(f"\n[CORRECTION] Applying corrections...")
    
    # Downgrade trust score based on violations
    validation_penalty = state['validation_results']['validation_penalty']
    
    # For demo: just apply the penalty (in production, retry extraction)
    state['iteration_count'] += 1
    
    # Force exit after correction attempt
    state['validation_results']['violations'] = []  # Clear to exit loop
    
    print(f"  Applied validation penalty: {validation_penalty} points")
    return state

# Final scoring node
def finalize_scoring(state: FacilityValidationState) -> FacilityValidationState:
    """
    Calculate final trust score incorporating validation penalties.
    """
    print(f"\n[FINALIZE] Calculating final trust score...")
    
    base_penalty = state['raw_data'].get('Trust_Penalty', 0)
    validation_penalty = state['validation_results']['validation_penalty']
    
    total_penalty = min(base_penalty + validation_penalty, 100)
    final_score = 100 - total_penalty
    
    # Build trust reason with citations
    violations = state['validation_results']['violations']
    trust_reason_parts = []
    
    for violation in violations:
        if 'specialty' in violation:
            citation = f"[Doc: {state['source_document_id']}, Page 1]"
            trust_reason_parts.append(
                f"Missing {violation['missing_requirement']} for {violation['specialty']} {citation}"
            )
    
    state['trust_score'] = final_score
    state['trust_reason'] = '; '.join(trust_reason_parts) if trust_reason_parts else 'Passed validation'
    
    print(f"  Final Trust Score: {final_score}/100")
    return state

# Build the LangGraph workflow
workflow = StateGraph(FacilityValidationState)

# Add nodes
workflow.add_node("extract", extractor_agent)
workflow.add_node("validate", validator_agent)
workflow.add_node("correct", correction_agent)
workflow.add_node("finalize", finalize_scoring)

# Define edges
workflow.set_entry_point("extract")
workflow.add_edge("extract", "validate")
workflow.add_conditional_edges(
    "validate",
    should_correct,
    {
        "correct": "correct",
        "finalize": "finalize"
    }
)
workflow.add_edge("correct", "validate")  # Loop back
workflow.add_edge("finalize", END)

# Compile the graph
app = workflow.compile()

print("\n" + "="*70)
print("Multi-Agent Validation System Ready")
print("="*70)
print("\nWorkflow: Extractor → Validator → [Correction Loop] → Finalize")
print("\nMLflow Tracing: Enabled - view in Databricks UI")
````

````python
%sql
-- CRITICAL ENHANCEMENT 4: NGO Actionable Insights Dashboard
-- Addresses Social Impact (25%) - Clear, actionable recommendations

-- Create NGO Planning Dashboard View
CREATE OR REPLACE VIEW ngo_medical_desert_action_plan AS
WITH regional_analysis AS (
    SELECT 
        h3_index_res6 AS region_id,
        FIRST(address_stateOrRegion) AS state,
        FIRST(address_country) AS country,
        COUNT(*) AS total_facilities,
        COUNT(CASE WHEN Trust_Score >= 70 THEN 1 END) AS trusted_facilities,
        COUNT(CASE WHEN facilityTypeId = 'hospital' THEN 1 END) AS hospitals,
        COUNT(CASE WHEN specialties LIKE '%surgery%' OR specialties LIKE '%Surgery%' THEN 1 END) AS surgical_centers,
        COUNT(CASE WHEN specialties LIKE '%emergency%' OR specialties LIKE '%Emergency%' THEN 1 END) AS emergency_centers,
        COUNT(CASE WHEN specialties LIKE '%pediatric%' OR specialties LIKE '%Pediatric%' THEN 1 END) AS pediatric_centers,
        COUNT(CASE WHEN specialties LIKE '%trauma%' OR specialties LIKE '%Trauma%' THEN 1 END) AS trauma_centers,
        -- Estimate population served (rough estimate: 1 facility per 5000 people)
        COUNT(*) * 5000 AS estimated_population_served,
        AVG(Trust_Score) AS avg_trust_score
    FROM gold_healthcare_facilities_geo
    WHERE h3_index_res6 IS NOT NULL
    GROUP BY h3_index_res6
),
gap_analysis AS (
    SELECT 
        *,
        CASE 
            WHEN hospitals = 0 THEN 'CRITICAL: No hospitals in region'
            WHEN emergency_centers = 0 THEN 'HIGH: No emergency care facilities'
            WHEN surgical_centers = 0 THEN 'HIGH: No surgical capabilities'
            WHEN pediatric_centers = 0 THEN 'MEDIUM: No dedicated pediatric care'
            WHEN trusted_facilities = 0 THEN 'HIGH: Zero trusted facilities'
            WHEN avg_trust_score < 50 THEN 'MEDIUM: Low trust facilities only'
            ELSE 'ADEQUATE'
        END AS primary_gap,
        CASE 
            WHEN hospitals = 0 THEN 100
            WHEN emergency_centers = 0 THEN 90
            WHEN surgical_centers = 0 THEN 80
            WHEN pediatric_centers = 0 THEN 60
            WHEN trusted_facilities = 0 THEN 70
            WHEN avg_trust_score < 50 THEN 50
            ELSE 10
        END AS intervention_priority_score,
        -- Recommended interventions
        CASE 
            WHEN hospitals = 0 THEN 'Build or upgrade existing clinic to hospital status with 24/7 emergency services'
            WHEN emergency_centers = 0 THEN 'Establish emergency response unit; train staff in trauma care; acquire ambulance'
            WHEN surgical_centers = 0 THEN 'Partner with existing facilities to add surgical capability; hire anesthesiologist'
            WHEN pediatric_centers = 0 THEN 'Add pediatrician; acquire age-appropriate equipment; vaccination cold chain'
            WHEN trusted_facilities = 0 THEN 'Quality improvement program; staff training; equipment audit and upgrade'
            ELSE 'Maintain and monitor existing services'
        END AS recommended_intervention,
        -- Estimated investment needed (in USD)
        CASE 
            WHEN hospitals = 0 THEN 500000
            WHEN emergency_centers = 0 THEN 150000
            WHEN surgical_centers = 0 THEN 200000
            WHEN pediatric_centers = 0 THEN 75000
            WHEN trusted_facilities = 0 THEN 50000
            ELSE 10000
        END AS estimated_investment_usd
    FROM regional_analysis
)
SELECT 
    region_id,
    state,
    country,
    total_facilities,
    trusted_facilities,
    hospitals,
    surgical_centers,
    emergency_centers,
    pediatric_centers,
    trauma_centers,
    estimated_population_served,
    ROUND(avg_trust_score, 1) AS avg_trust_score,
    primary_gap,
    intervention_priority_score,
    recommended_intervention,
    estimated_investment_usd,
    -- Implementation timeline recommendation
    CASE 
        WHEN intervention_priority_score >= 90 THEN 'IMMEDIATE (0-3 months)'
        WHEN intervention_priority_score >= 70 THEN 'HIGH PRIORITY (3-12 months)'
        WHEN intervention_priority_score >= 50 THEN 'MEDIUM PRIORITY (1-2 years)'
        ELSE 'ONGOING MONITORING'
    END AS implementation_timeline
FROM gap_analysis
WHERE intervention_priority_score > 10
ORDER BY intervention_priority_score DESC, estimated_population_served DESC;

-- Show top 20 priority regions for NGO intervention
SELECT 
    state,
    country,
    total_facilities,
    hospitals,
    emergency_centers,
    estimated_population_served,
    primary_gap,
    recommended_intervention,
    estimated_investment_usd,
    implementation_timeline
FROM ngo_medical_desert_action_plan
LIMIT 20;
````

````python
%sql
-- Create optimized table for API queries
-- Supports search by pincode, city, geospatial coordinates, and trust score

CREATE OR REPLACE TABLE api_facilities_for_map AS
SELECT 
    -- Facility identification
    name AS facility_name,
    facilityTypeId AS facility_type,
    
    -- Location data for map plotting
    CAST(latitude AS DOUBLE) AS latitude,
    CAST(longitude AS DOUBLE) AS longitude,
    address_line1,
    address_line2,
    address_city AS city,
    address_stateOrRegion AS state,
    address_zipOrPostcode AS pincode,
    address_country AS country,
    
    -- H3 indices for fast spatial queries
    h3_index_res8,
    h3_index_res7,
    h3_index_res6,
    
    -- Trust and quality metrics
    Trust_Score AS trust_score,
    Trust_Penalty AS trust_penalty,
    Trust_Reason AS trust_reason,
    CASE 
        WHEN Trust_Score >= 90 THEN 'Highly Trusted'
        WHEN Trust_Score >= 70 THEN 'Trusted'
        WHEN Trust_Score >= 50 THEN 'Moderate Risk'
        ELSE 'High Risk'
    END AS risk_category,
    
    -- Contact information
    officialPhone AS phone,
    email,
    officialWebsite AS website,
    
    -- Services and capabilities
    specialties,
    equipment,
    capability,
    numberDoctors AS doctor_count,
    
    -- Social media presence (for trust indicators)
    CASE WHEN facebookLink IS NOT NULL AND facebookLink != 'null' THEN true ELSE false END AS has_facebook,
    CASE WHEN twitterLink IS NOT NULL AND twitterLink != 'null' THEN true ELSE false END AS has_twitter,
    CASE WHEN linkedinLink IS NOT NULL AND linkedinLink != 'null' THEN true ELSE false END AS has_linkedin,
    
    -- Metadata for frontend display
    description,
    yearEstablished AS year_established,
    distinct_social_media_presence_count AS social_media_count,
    
    -- Color coding for map markers (hex colors)
    CASE 
        WHEN Trust_Score >= 90 THEN '#10B981'  -- Green
        WHEN Trust_Score >= 70 THEN '#3B82F6'  -- Blue
        WHEN Trust_Score >= 50 THEN '#F59E0B'  -- Amber
        ELSE '#EF4444'  -- Red
    END AS map_marker_color
FROM gold_healthcare_facilities_geo
WHERE latitude IS NOT NULL 
  AND longitude IS NOT NULL
  AND latitude != 'null'
  AND longitude != 'null';

-- Note: CREATE INDEX is not supported in Unity Catalog
-- Delta Lake uses Liquid Clustering and Z-ordering for optimization instead
-- For production, consider: OPTIMIZE api_facilities_for_map ZORDER BY (city, pincode, h3_index_res7)

-- Test query: Search by city
SELECT COUNT(*) as total_facilities FROM api_facilities_for_map WHERE city = 'Delhi';
SELECT 'API table created successfully!' AS status;
````

````python
# Healthcare Facilities REST API
# FastAPI service for map frontend integration

from fastapi import FastAPI, HTTPException, Query
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
from typing import List, Optional
import json

# Initialize FastAPI app
app = FastAPI(
    title="Healthcare Facilities API",
    description="REST API for healthcare facility discovery with trust scoring and geospatial search",
    version="1.0.0"
)

# Enable CORS for frontend integration
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # In production, specify your frontend domain
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Response models
class FacilityResponse(BaseModel):
    facility_name: str
    facility_type: Optional[str]
    latitude: float
    longitude: float
    city: Optional[str]
    state: Optional[str]
    pincode: Optional[str]
    trust_score: int
    risk_category: str
    map_marker_color: str
    phone: Optional[str]
    website: Optional[str]
    specialties: Optional[str]
    doctor_count: Optional[str]
    trust_reason: Optional[str]

class FacilityDetailResponse(FacilityResponse):
    address_line1: Optional[str]
    address_line2: Optional[str]
    email: Optional[str]
    equipment: Optional[str]
    capability: Optional[str]
    description: Optional[str]
    year_established: Optional[str]
    has_facebook: bool
    has_twitter: bool
    has_linkedin: bool
    social_media_count: Optional[str]

class SearchResponse(BaseModel):
    total_count: int
    facilities: List[FacilityResponse]

# Helper function to execute SQL queries
def execute_query(query: str) -> List[dict]:
    """Execute SQL query and return results as list of dicts"""
    try:
        result_df = spark.sql(query)
        # Convert to pandas then to dict for JSON serialization
        pandas_df = result_df.toPandas()
        return pandas_df.to_dict(orient='records')
    except Exception as e:
        raise HTTPException(status_code=500, detail=f"Database error: {str(e)}")

# API Endpoints

@app.get("/", tags=["Health Check"])
def root():
    """API health check endpoint"""
    return {
        "status": "healthy",
        "service": "Healthcare Facilities API",
        "version": "1.0.0",
        "documentation": "/docs"
    }

@app.get("/api/v1/facilities/search/city", response_model=SearchResponse, tags=["Search"])
def search_by_city(
    city: str = Query(..., description="City name to search"),
    limit: int = Query(100, ge=1, le=1000, description="Maximum number of results")
):
    """
    Search facilities by city name.
    
    Example: /api/v1/facilities/search/city?city=Delhi&limit=50
    """
    query = f"""
        SELECT 
            facility_name, facility_type, latitude, longitude,
            city, state, pincode, trust_score, risk_category,
            map_marker_color, phone, website, specialties,
            doctor_count, trust_reason
        FROM api_facilities_for_map
        WHERE LOWER(city) = LOWER('{city}')
        ORDER BY trust_score DESC
        LIMIT {limit}
    """
    
    results = execute_query(query)
    return {
        "total_count": len(results),
        "facilities": results
    }

@app.get("/api/v1/facilities/search/pincode", response_model=SearchResponse, tags=["Search"])
def search_by_pincode(
    pincode: str = Query(..., description="Pincode/ZIP code to search"),
    limit: int = Query(100, ge=1, le=1000, description="Maximum number of results")
):
    """
    Search facilities by pincode.
    
    Example: /api/v1/facilities/search/pincode?pincode=110001&limit=50
    """
    query = f"""
        SELECT 
            facility_name, facility_type, latitude, longitude,
            city, state, pincode, trust_score, risk_category,
            map_marker_color, phone, website, specialties,
            doctor_count, trust_reason
        FROM api_facilities_for_map
        WHERE pincode = '{pincode}'
        ORDER BY trust_score DESC
        LIMIT {limit}
    """
    
    results = execute_query(query)
    return {
        "total_count": len(results),
        "facilities": results
    }

@app.get("/api/v1/facilities/search/bounds", response_model=SearchResponse, tags=["Search"])
def search_by_bounds(
    min_lat: float = Query(..., description="Minimum latitude"),
    max_lat: float = Query(..., description="Maximum latitude"),
    min_lon: float = Query(..., description="Minimum longitude"),
    max_lon: float = Query(..., description="Maximum longitude"),
    min_trust_score: int = Query(0, ge=0, le=100, description="Minimum trust score filter"),
    limit: int = Query(500, ge=1, le=2000, description="Maximum number of results")
):
    """
    Search facilities within map viewport bounds.
    Optimized for map rendering - returns facilities visible in current map view.
    
    Example: /api/v1/facilities/search/bounds?min_lat=28.4&max_lat=28.8&min_lon=77.0&max_lon=77.4&limit=200
    """
    query = f"""
        SELECT 
            facility_name, facility_type, latitude, longitude,
            city, state, pincode, trust_score, risk_category,
            map_marker_color, phone, website, specialties,
            doctor_count, trust_reason
        FROM api_facilities_for_map
        WHERE latitude BETWEEN {min_lat} AND {max_lat}
          AND longitude BETWEEN {min_lon} AND {max_lon}
          AND trust_score >= {min_trust_score}
        ORDER BY trust_score DESC
        LIMIT {limit}
    """
    
    results = execute_query(query)
    return {
        "total_count": len(results),
        "facilities": results
    }

@app.get("/api/v1/facilities/search/radius", response_model=SearchResponse, tags=["Search"])
def search_by_radius(
    lat: float = Query(..., description="Center latitude"),
    lon: float = Query(..., description="Center longitude"),
    radius_km: float = Query(10, ge=0.1, le=100, description="Search radius in kilometers"),
    min_trust_score: int = Query(0, ge=0, le=100, description="Minimum trust score filter"),
    limit: int = Query(100, ge=1, le=1000, description="Maximum number of results")
):
    """
    Search facilities within radius of a point using H3 spatial indexing.
    Ultra-fast proximity search.
    
    Example: /api/v1/facilities/search/radius?lat=28.6139&lon=77.2090&radius_km=5&limit=50
    """
    # Use H3 for fast spatial search
    query = f"""
        WITH target_h3 AS (
            SELECT h3_longlatash3({lon}, {lat}, 7) AS target_cell
        ),
        nearby_cells AS (
            SELECT EXPLODE(h3_kring(target_cell, 2)) AS cell FROM target_h3
        ),
        candidate_facilities AS (
            SELECT 
                f.facility_name, f.facility_type, f.latitude, f.longitude,
                f.city, f.state, f.pincode, f.trust_score, f.risk_category,
                f.map_marker_color, f.phone, f.website, f.specialties,
                f.doctor_count, f.trust_reason,
                ST_DISTANCE(
                    ST_POINT({lon}, {lat}),
                    ST_POINT(f.longitude, f.latitude)
                ) / 1000 AS distance_km
            FROM api_facilities_for_map f
            INNER JOIN nearby_cells nc ON f.h3_index_res7 = nc.cell
            WHERE f.trust_score >= {min_trust_score}
        )
        SELECT 
            facility_name, facility_type, latitude, longitude,
            city, state, pincode, trust_score, risk_category,
            map_marker_color, phone, website, specialties,
            doctor_count, trust_reason
        FROM candidate_facilities
        WHERE distance_km <= {radius_km}
        ORDER BY distance_km ASC, trust_score DESC
        LIMIT {limit}
    """
    
    results = execute_query(query)
    return {
        "total_count": len(results),
        "facilities": results
    }

@app.get("/api/v1/facilities/detail/{facility_name}", response_model=FacilityDetailResponse, tags=["Details"])
def get_facility_detail(facility_name: str):
    """
    Get complete details for a specific facility.
    
    Example: /api/v1/facilities/detail/32%20Kings%20Dental%20Clinic
    """
    # Escape single quotes in facility name
    safe_name = facility_name.replace("'", "''")
    
    query = f"""
        SELECT *
        FROM api_facilities_for_map
        WHERE facility_name = '{safe_name}'
        LIMIT 1
    """
    
    results = execute_query(query)
    
    if not results:
        raise HTTPException(status_code=404, detail="Facility not found")
    
    return results[0]

@app.get("/api/v1/facilities/stats", tags=["Statistics"])
def get_statistics():
    """
    Get overall statistics about facilities in the database.
    """
    query = """
        SELECT 
            COUNT(*) AS total_facilities,
            COUNT(DISTINCT city) AS unique_cities,
            COUNT(DISTINCT state) AS unique_states,
            COUNT(DISTINCT pincode) AS unique_pincodes,
            ROUND(AVG(trust_score), 1) AS avg_trust_score,
            COUNT(CASE WHEN trust_score >= 90 THEN 1 END) AS highly_trusted_count,
            COUNT(CASE WHEN trust_score >= 70 THEN 1 END) AS trusted_count,
            COUNT(CASE WHEN trust_score < 50 THEN 1 END) AS high_risk_count
        FROM api_facilities_for_map
    """
    
    results = execute_query(query)
    return results[0] if results else {}

@app.get("/api/v1/facilities/cities", tags=["Utilities"])
def list_cities(limit: int = Query(100, ge=1, le=500)):
    """
    Get list of all cities with facility counts.
    Useful for frontend city dropdown/autocomplete.
    """
    query = f"""
        SELECT 
            city,
            state,
            COUNT(*) AS facility_count,
            ROUND(AVG(trust_score), 1) AS avg_trust_score
        FROM api_facilities_for_map
        WHERE city IS NOT NULL
        GROUP BY city, state
        ORDER BY facility_count DESC
        LIMIT {limit}
    """
    
    results = execute_query(query)
    return {"cities": results}

print("\n" + "="*70)
print("  FastAPI Application Configured")
print("="*70)
print("\nAPI endpoints created successfully!")
print("\nAvailable endpoints:")
print("  GET  /                                    - Health check")
print("  GET  /api/v1/facilities/search/city      - Search by city name")
print("  GET  /api/v1/facilities/search/pincode   - Search by pincode")
print("  GET  /api/v1/facilities/search/bounds    - Search within map bounds")
print("  GET  /api/v1/facilities/search/radius    - Search within radius (H3 optimized)")
print("  GET  /api/v1/facilities/detail/{name}    - Get facility details")
print("  GET  /api/v1/facilities/stats            - Get overall statistics")
print("  GET  /api/v1/facilities/cities           - List all cities")
print("\nNext: Run the API server (see next cell)")
print("="*70)
````
